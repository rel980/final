# Input Data Siswa

```python
# List untuk menyimpan data siswa
data_siswa = []

# Fungsi untuk menambahkan data siswa
def tambah_siswa():
    print("\nMasukkan data siswa:")
    nama = input("Nama: ")
    nim = input("NIM: ")
    jurusan = input("Jurusan: ")
    try:
        nilai = float(input("Nilai: "))
    except ValueError:
        print("Nilai harus berupa angka.")
        return

    # Membuat dictionary untuk satu siswa
    siswa = {
        "Nama": nama,
        "NIM": nim,
        "Jurusan": jurusan,
        "Nilai": nilai
    }

    # Menambahkan dictionary ke dalam list
    data_siswa.append(siswa)
    print("\nData siswa berhasil ditambahkan!\n")

# Fungsi untuk menampilkan semua data siswa
def tampilkan_data_siswa():
    if not data_siswa:
        print("\nBelum ada data siswa.\n")
        return

    print("\nData Siswa:")
    for idx, siswa in enumerate(data_siswa, start=1):
        print(f"Siswa {idx}:")
        print(f"  Nama     : {siswa['Nama']}")
        print(f"  NIM      : {siswa['NIM']}")
        print(f"  Jurusan  : {siswa['Jurusan']}")
        print(f"  Nilai    : {siswa['Nilai']:.2f}")
        print()

# Fungsi untuk mencari siswa berdasarkan NIM
def cari_siswa():
    if not data_siswa:
        print("\nBelum ada data siswa.\n")
        return

    nim_cari = input("\nMasukkan NIM yang ingin dicari: ")
    for siswa in data_siswa:
        if siswa["NIM"] == nim_cari:
            print("\nData Siswa:")
            print(f"  Nama     : {siswa['Nama']}")
            print(f"  NIM      : {siswa['NIM']}")
            print(f"  Jurusan  : {siswa['Jurusan']}")
            print(f"  Nilai    : {siswa['Nilai']:.2f}")
            return
    print("\nSiswa dengan NIM tersebut tidak ditemukan.\n")

# Fungsi untuk menghitung rata-rata nilai
def hitung_rata_rata():
    if not data_siswa:
        print("\nBelum ada data siswa.\n")
        return

    total_nilai = sum(siswa["Nilai"] for siswa in data_siswa)
    rata_rata = total_nilai / len(data_siswa)
    print(f"\nRata-rata nilai dari semua siswa adalah: {rata_rata:.2f}\n")

# Menu utama
while True:
    print("Menu:")
    print("1. Tambah Data Siswa")
    print("2. Tampilkan Semua Data Siswa")
    print("3. Cari Siswa Berdasarkan NIM")
    print("4. Hitung Rata-rata Nilai")
    print("5. Keluar")

    pilihan = input("Pilih menu (1/2/3/4/5): ")

    if pilihan == "1":
        tambah_siswa()
    elif pilihan == "2":
        tampilkan_data_siswa()
    elif pilihan == "3":
        cari_siswa()
    elif pilihan == "4":
        hitung_rata_rata()
    elif pilihan == "5":
        print("Keluar dari program.")
        break
    else:
        print("Pilihan tidak valid. Silakan coba lagi.\n")
```
